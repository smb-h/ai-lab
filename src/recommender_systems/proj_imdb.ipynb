در این پروژه شما باید یک سیستم توصیه‌گر بسازید . داده‌ها مربوط به فیلم های وبسایت IMDB هستند. دو فایل وجود دارد که یکی لیست فیلم‌ها و ژانر آن‌ها و دیگری امتیاز کاربران به این فیلم‌ها. از آنجایی این مثال در آزمایشگاه توضیح داده شده است، انتظار می‌رود یک فایل نوتبوک جدید بسازید و کدها را توضیح دهید. نتیجه کدها و گزارش خود را در قالب جوپیتر نوتبوک ارسال کنید. هرچه گزارش شما و توضیح شما درباره کدها کامل تر باشد نمره کامل تری نیز دریافت می‌کنید. فایل‌های مورد نیاز به صورت زیپ در پیوست آمده‌اند.
