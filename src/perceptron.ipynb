```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.linear_model import Perceptron
```

```python
# array from 0 to 10 with 0.05 step
x = np.arange(0, 10, 0.05)
y = (0.001 * (x ** 2)) + x * np.sin(x)
```

```python
# plot x and y
plt.plot(x, y)
```

```python
# noise with size of x and fill it with random values (not normal)
noise = np.random.rand(len(x))
# reduce noise by its mean
noise = noise - np.mean(noise)
noise_y = y + noise
```

```python
# plot x,y and x, noise_y
plt.plot(x, y, "b")
plt.plot(x, noise_y, "r")
```

# Neural network

```python
digits = load_digits()
print(digits.data.shape)
```

```python
plt.gray()
plt.matshow(digits.images[100])
```

```python
x, y = load_digits(return_X_y=True)
print(x.shape)
display(x)
print(y.shape)
display(y)
```

```python
clf = Perceptron(tol=1e-3, random_state=0)
clf.fit(x, y)
clf.score(x, y)
```

```python
display(x[100])
display(x[100].shape)
clf.predict([x[100]])
```
