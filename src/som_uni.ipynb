```python
from sklearn_som.som import SOM
from matplotlib import pyplot as plt
from matplotlib.colors import ListedColormap
from numpy.random import rand, randn
from numpy import shape, sin, transpose, pi, arange, array
```

```python
data_points = 1000

x1 = 2 * pi * rand(data_points)
noise = 0.1 * randn(data_points)
x2 = sin(x1) + noise

x = transpose(array([x1, x2]))

plt.scatter(x1, x2, marker=".")
print("shape: ", shape(x))
```

```python
date_dim = 2
m = 10
model = SOM(m, n=1, dim=date_dim)
model.fit(x)

cluster_centers = model.cluster_centers_
cluster_centers = cluster_centers.reshape(-1, date_dim)

cluster_label = model.predict(x)
```

```python
colors = [
    "green",
    "brown",
    "orange",
    "gold",
    "blue",
    "black",
    "pink",
    "purple",
    "red",
    "#550011",
]
cmap = ListedColormap(colors)

plt.scatter(x1, x2, marker=".", c=cluster_label, cmap=cmap, alpha=0.2)

plt.plot(cluster_centers[:, 0], cluster_centers[:, 1], color="black")
plt.scatter(
    cluster_centers[:, 0], cluster_centers[:, 1], marker="*", color="black", s=700
)
plt.scatter(
    cluster_centers[:, 0],
    cluster_centers[:, 1],
    marker="*",
    edgecolor="white",
    s=400,
    c=arange(0, m),
    cmap=cmap,
)
```
